```python
# load sample data-set from sklearn load_iris
import pandas as pd
from sklearn.datasets import load_iris

iris_data = load_iris()
df = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
target = pd.Series(iris_data.target).map(
    {k: v for k, v in enumerate(iris_data.target_names)})
df["target"] = target
df
```

```python
import seaborn as sns

# make a scatter plot
sns.scatterplot(x="sepal length (cm)", y="sepal width (cm)", data=df)

# make a scatter plot with hue
sns.scatterplot(x="sepal length (cm)", y="sepal width (cm)", hue="target", data=df)

# make a scatter plot with hue and style
sns.scatterplot(x="sepal length (cm)", y="sepal width (cm)", hue="target", style="target", data=df)

# make a scatter plot with hue and style and size
sns.scatterplot(x="sepal length (cm)", y="sepal width (cm)", hue="target", style="target", size="petal length (cm)", data=df)
```

```python
# make a bar plot
sns.barplot(x="target", y="sepal length (cm)", data=df)

# make a bar plot with hue
sns.barplot(x="target", y="sepal length (cm)", hue="target", data=df)
```

```python
# make a box plot
# sns.boxplot(x="target", y="sepal length (cm)", data=df)

# make a box plot with hue
sns.boxplot(x="target", y="sepal length (cm)", hue="target", data=df)
```

```python
# make a violin plot
# sns.violinplot(x="target", y="sepal length (cm)", data=df)

# make a violin plot with hue
sns.violinplot(x="target", y="sepal length (cm)", hue="target", data=df)
```

```python
# make a swarm plot
# sns.swarmplot(x="target", y="sepal length (cm)", data=df)

# make a swarm plot with hue
sns.swarmplot(x="target", y="sepal length (cm)", hue="target", data=df)
```
